```r
library(bigrquery)
library(httr)

run_bq_query <- function(query, project_id = "cal-itp-data-infra") {
  # Get access token from gcloud
  access_token <- system("gcloud auth application-default print-access-token", intern = TRUE)

  # Construct token object
  token <- Token2.0$new(
    app = oauth_app("dummy", key = "", secret = ""),
    endpoint = NULL,
    credentials = list(access_token = access_token),
    params = list(as_header = TRUE)
  )

  # Authenticate with bigrquery
  bq_auth(token = token)

  # Run the query and return the result as a data frame
  bq_table_download(bq_project_query(project_id, query))
}
```

```r
run_bq_query("SELECT * FROM `cal-itp-data-infra.mart_transit_database.dim_service_components` LIMIT 5")
```
