```python
#https://medium.com/analytics-vidhya/create-and-visualize-choropleth-map-with-folium-269d3fd12fa0
import geopandas as gpd
import intake
import pandas as pd
import branca
import folium
```

```python
#Must put in 2 arguments, location & zoom. Everything else is 'optional.'
sf = (folium.Map(location=[37.7790262,-122.419906], 
                 tiles= "Stamen Toner", zoom_start=12, #smaller the #, the more the map will be zoomed out.
                 width=500, height=500, control_scale= True))
sf
```

```python
sf = folium.Map(
    location=[37.7790262,-122.41990],
    zoom_start=12)
folium.Marker([37.802379,-122.4058344], popup='Coit Tower', icon=folium.Icon(icon='cloud', color ='pink'),tooltip="Coit Tower is here").add_to(sf)
folium.Marker([37.7714804,-122.4691275],popup='De Young Museum is over here',icon=folium.Icon(color='purple')).add_to(sf)
sf
```
