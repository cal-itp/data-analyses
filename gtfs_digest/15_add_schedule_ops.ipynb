### Editing `_operators_prep`

```python
schd_vp_url = f"{GTFS_DATA_DICT.digest_tables.dir}{GTFS_DATA_DICT.digest_tables.route_schedule_vp}.parquet"
```

```python
schd_vp_df = pd.read_parquet(schd_vp_url)
```

```python
schd_vp_df.sched_rt_category.unique()
```

```python
schd_vp_df = pd.read_parquet(
    schd_vp_url,
    filters=[[("sched_rt_category", "in", ["schedule_and_vp", "schedule_only"])]],
    columns=[
        "schedule_gtfs_dataset_key",
        "caltrans_district",
        "organization_name",
        "name",
        "sched_rt_category",
        "service_date",
    ],
)
```

```python
(schd_vp_df
                      .dropna(subset="caltrans_district")
                      .sort_values(by=["caltrans_district", "organization_name", "service_date", "sched_rt_category"],
            ascending=[True, True, False, False])
                      .drop_duplicates(subset = ["schedule_gtfs_dataset_key", "caltrans_district",])
                      .drop_duplicates(subset = ["caltrans_district", "organization_name"])
                     .drop_duplicates(subset = ["caltrans_district", "name"])
                      .reset_index(drop = True)
                     )
```
