```python
def find_late(mart_table:str, date_column:str):
    df = query_sql(f"""
    SELECT *
    FROM mart_payments.{mart_table}
    WHERE {date_column} > "2026-04-06"
    """)
    print("length of dataframe")
    print(len(df))
    mask = df.astype(str).apply(lambda col: col.str.contains("late", case=False, na=False))
    df2 = df[mask.any(axis=1)]
    display(df2.T)
  
```
