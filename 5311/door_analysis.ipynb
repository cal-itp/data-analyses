```python
df_agg = df.groupby(['organization_name','reporter_type','GTFS', 'fleet_size']).agg({'total_vehicles':'max',
       'average_age_of_fleet__in_years_':'max','doors_sum':'max','_0_9':'max', '_10_12':'max', '_13_15':'max', '_16_20':'max', '_21_25':'max',
       '_26_30':'max', '_31_60':'max', '_60plus':'max',})
```

```python
df_agg= df_agg.reset_index()
```

```python
column_names = ['_16_20', '_21_25', '_26_30', '_31_60']
df_agg['_16plus']= df_agg[column_names].sum(axis=1)
```
