```python
import pandas as pd
import utils
```

```python
dim_trips = pd.read_parquet(
    "dim_trips.parquet",
    filters = [[
        ("feed_key", "in", list(utils.feed_to_name_dict.keys())),
    ]],
    columns = ["feed_key", "route_id", "service_id", "trip_id"]
).pipe(utils.get_operator_lax_routes)

dim_calendar = pd.read_parquet(
    "dim_calendar.parquet",
    filters = [[("feed_key", "in", utils.subset_feeds)]],
    columns = ["feed_key", "service_id", "start_date", "end_date"]
)

df = pd.merge(
    dim_calendar,
    dim_trips,
    on = ["feed_key", "service_id"],
    how = "inner"
)
```

```python
results = (
    df.groupby(["gtfs_dataset_name", "route_name"])
    .agg({
        "start_date": lambda x: list(set(x)),
        "end_date": lambda x: list(set(x))
    }).reset_index()
)

results
```
