# 📍 Transit Stop Analysis Tool

This notebook identifies **bus and rail stops** near any venue/location and displays interactive maps and summary tables.

### How to use:
1. From the top menu click **Run** → **Run All Cells**
2. A text box and a button will appear below
3. Optionally adjust the buffer distances
4. Click **▶ Run Analysis**

> ⚠️ Please wait for the maps and tables to fully load before entering a new venue.

```python
import sys
sys.path.insert(0, '.')

import ipywidgets as widgets
from IPython.display import display, clear_output
from utils.geo_functions import run_analysis

# --- Input widgets ---
venue_input = widgets.Text(
    value='',
    placeholder="e.g. SoFi Stadium, Levi's Stadium",
    description='📍 Venue/Location:',
    layout=widgets.Layout(width='420px')
)

bus_buffer_input = widgets.FloatSlider(
    value=3,
    min=0.5, max=10, step=0.5,
    description='🚌 Bus buffer (mi):',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='420px')
)

rail_buffer_input = widgets.FloatSlider(
    value=10,
    min=1, max=30, step=1,
    description='🚆 Rail buffer (mi):',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='420px')
)

run_button = widgets.Button(
    description='▶ Run Analysis',
    button_style='success',
    layout=widgets.Layout(width='160px', margin='10px 0 0 0')
)

output_area = widgets.Output()

# --- Button logic ---
def on_run_clicked(b):
    venue = venue_input.value.strip()
    with output_area:
        clear_output(wait=True)
        if not venue:
            print('⚠️  Please enter a venue name first.')
            return
        run_analysis(
            venue_name=venue,
            bus_buffer_miles=bus_buffer_input.value,
            rail_buffer_miles=rail_buffer_input.value
        )

run_button.on_click(on_run_clicked)

# --- Display ---
display(
    venue_input,
    bus_buffer_input,
    rail_buffer_input,
    run_button,
    output_area
)
```
