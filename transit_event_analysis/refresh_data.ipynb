# 🔄 Refresh Stops Data

Run this notebook **only when you need to fetch fresh transit stops data from BigQuery.**

This will overwrite the existing `df_dim_stops_latest.parquet` file in the root directory.

> ⚠️ This query bills against BigQuery — do not run unnecessarily.

### How to use:
1. From the top menu click **Run** → **Run All Cells**
2. Wait for the query to complete — this may take a few minutes
3. Once done, the updated parquet file will be ready for use in `analysis.ipynb`

```python
from google.cloud import bigquery
from calitp_data_analysis.sql import get_engine
from pathlib import Path
import pandas as pd

# Load the SQL query from the query file
query = Path('query.sql').read_text()

# Run the BigQuery query
print('⏳ Fetching data from BigQuery, please wait...')
job_config = bigquery.QueryJobConfig(maximum_bytes_billed=2_000_000_000_000)
engine = get_engine()
df_dim_stops_latest = pd.read_sql_query(
    query,
    engine.execution_options(job_config=job_config)
)

# Save as parquet in the root directory
df_dim_stops_latest.to_parquet('df_dim_stops_latest.parquet', index=False)
print(f'✅ Done! {len(df_dim_stops_latest):,} rows saved to df_dim_stops_latest.parquet')
df_dim_stops_latest.head()
```
