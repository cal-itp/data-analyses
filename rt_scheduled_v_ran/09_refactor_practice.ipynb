# Refactor for Performance

1. Start with this [metrics](https://github.com/cal-itp/data-analyses/blob/main/rt_scheduled_v_ran/scripts/metrics.md) file
2. Notice any broad categorizations of how the metrics would be generated
   * is it spatial or tabular (strictly aggregation)
   * is it by trip or first trip-minute, then trip?
   * note: aggregations by trip or trip-minute are fairly quick and can be handled by `dask`, even if it's just `dask.delayed`. `pd.Grouper(time)` must be `pandas`.
3. Try to keep the categorizations together and write a function for each
   * It's ok if the metric cannot be fully generated (aka, you only calculate a numerator or denominator)
   * Note how long it takes for each portion
   * If one function takes a very long time, it might be worth it to investigate writing it differently
4. Put it together
5. Aggregate it further and see how well you can reuse functions
   * We have a trip aggregation already
   * Then we have a route-direction aggregation
   * If you have a function that generates the metrics, can you use it for both trip and route-direction? If not, rewrite it, separate out what's giving you an error, etc.
   * If there are things that should be widely used, move to a module like `segment_speed_utils`
