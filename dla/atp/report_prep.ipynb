## Merging SUR Funding and Details

* merging on `project_app_id` instead of `atp_id`

```python
#columns in common between sur_details and sur_funding
sur_funding.columns.intersection(sur_details.columns)
```

```python
merge_on = ['project_app_id', 'project_cycle', 'a2_ct_dist', 'a1_locode']
```

```python
funded = (pd.merge(sur_details, sur_funding, how="outer", on = merge_on, indicator=True))
```

```python
funded._merge.value_counts()
```

```python
funded>>filter(_._merge=='right_only')
```

* rows that did not merge appear to be a total column and one Caltrans project that did not have a match. 

```python
#for now keeping full records
funded = funded>>filter(_._merge=='both')
```

```python
compare_entries = np.where(funded["a2_info_proj_name_x"] == funded["a2_info_proj_name_y"], True, False)
funded["compare_desc"] = compare_entries
```

```python
funded.compare_desc.value_counts()
```

```python
funded>>filter(_.compare_desc==False)>>select(_.a2_info_proj_name_x,  _.a2_info_proj_name_y)
```

```python
# filling the null values for some of the duplicate columns
# manually checking that values are the same as of now- will add function to check when we get the data links

funded['awarded_x'] = funded['awarded_x'].fillna(funded['awarded_y'])
funded['ppno_y'] = funded['ppno_y'].fillna(funded['ppno_x'])
```

```python
#renaming and dropping duplicate columns 
## a1_imp_agcy_name_x has manual errors so selecting a1_imp_agcy_name_y

funded = funded.rename(columns={'awarded_x':'awarded',
                                'ppno_y':'ppno',
                                'a1_imp_agcy_name_y':'a1_imp_agcy_name',
                                'a2_info_proj_name_y':'a2_info_proj_name'
                               })
funded = funded.drop(columns={'awarded_y', 'a1_imp_agcy_name_x', 'a2_info_proj_name_x','ppno_x', 'compare_desc'})
```
