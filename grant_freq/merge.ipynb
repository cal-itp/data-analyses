# Merging DLA and DRMT

```python
import pandas as pd
from siuba import *
import numpy as np
from IPython.display import Markdown
```

```python
pd.set_option("display.max_columns", 100)
pd.set_option("display.max_colwidth", 1000)
```

```python
GCS_FILE_PATH_DLA = 'gs://calitp-analytics-data/data-analyses/dla/e-76Obligated/'
```

```python
dla = pd.read_parquet(f'{GCS_FILE_PATH_DLA}dla_df.parquet')
```

```python
drmt = pd.read_parquet('drmt_grants.parquet')
```

```python
dla1 = dla>>count(_.primary_agency_name)
```

```python
dla1= dla1.rename(columns={'primary_agency_name':'organization_name'})
```

```python
drmt1 = drmt>>group_by(_.organization_name)>> summarize(n=_.grant_number.nunique())
```

```python
drmt1 = drmt1.rename(columns={'n':'nunique_grantnum'})
```

```python
## using Organization name and Primary Agency Name to match
```

```python
df_all = pd.merge(dla1, drmt1, on='organization_name', how='outer', indicator=True)
```

```python
df_all._merge.value_counts()
```
